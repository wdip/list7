```python
import random
import matplotlib.pyplot as plt
```

```python
class Disk():
    def __init__(self, x=0, y=0, r=1):
        self.x = x
        self.y = y
        self.r = r
        
    def __eq__(self, other):
        return (self.x == other.x and self.y == other.y and self.r == other.r)
    
    def move(self, vector):
        self.x += vector[0]
        self.y += vector[1]
        
    def get_distance(self, other):
        return ((self.x-other.x)**2 + (self.y-other.y)**2)**(1/2)
    
    def if_collision(self, other):
        return self.get_distance(other) < self.r + other.r
```

```python
disks = []
for i in range(100):
    disks.append(Disk(random.uniform(-14.5, 14.5), random.uniform(-14.5, 14.5), 0.5))
```

```python
for disk in disks:
    print(disk.x, disk.y, disk.r)
```

```python
plt.figure(figsize=(16, 9))
for disk in disks:
    circle = plt.Circle((disk.x, disk.y), disk.r)
    plt.gca().add_artist(circle)

plt.xlim((-15, 15))
plt.ylim((-15, 15))
plt.show()
```

```python
def if_collision_exists(disks):
    collided = False
    for disk in disks:
        for disk1 in disks:
            if disk.if_collision(disk1) and disk != disk1:
                collided = True
                
    return collided
                
def remove_collisions(disks):
    for disk in disks:
        for disk1 in disks:
            if disk.if_collision(disk1) and disk != disk1:
                disk.move([random.uniform(-1, 1), random.uniform(-1, 1)])
```

```python
while if_collision_exists(disks):
    remove_collisions(disks)
```

```python
plt.figure(figsize=(16, 9))
for disk in disks:
    circle = plt.Circle((disk.x, disk.y), disk.r)
    plt.gca().add_artist(circle)

plt.xlim((-15, 15))
plt.ylim((-15, 15))
plt.show()
```
